```python
from sklearn.ensemble import AdaBoostClassifier
def get_accuracy(clf):
    clf = clf()
    clf = clf.fit(train_x, train_y)
    y_pred = clf.predict(test_x)
    return (str(sum(test_y == y_pred)/test_y.shape[0]))

print('SVM: ', get_accuracy(SVC))
print('DecisionTree: ', get_accuracy(DecisionTreeClassifier))
print('RandomForest: ', get_accuracy(RandomForestClassifier))
print('AdaBoost: ', get_accuracy(AdaBoostClassifier))  ## Boosting的演算法
print('XGB: ', get_accuracy(XGBClassifier))
```

```python
from xgboost import XGBClassifier
params = {}
params['objective'] = 'binary:logistic'
params['eval_metric'] = 'logloss'
params['eta'] = 0.04
params['max_depth'] = 3
params['learning_rate'] = 0.01
```

```python
import xgboost
d_train = xgboost.DMatrix(train_x, label=train_y)
d_valid = xgboost.DMatrix(test_x, label=test_y)

watchlist = [(d_train, 'train'), (d_valid, 'valid')]

bst = xgboost.train(params, d_train, 100, watchlist, early_stopping_rounds=100, verbose_eval=10)
y_pred = bst.predict(xgboost.DMatrix(test_x))
print("Accuracy: ", str(sum(test_y == (y_pred > 0.5))/test_y.shape[0]))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'max_depth':range(3,10,2),  
              'min_child_weight':range(1,6,2),
              'reg_alpha': [0, 0.001, 0.005, 0.01, 0.05],
              'gamma': [i / 10.0 for i in range(0, 5)],
              'subsample': [i / 10.0 for i in range(6, 10)],  
               'colsample_bytree': [i / 10.0 for i in range(6, 10)] ,
              'n_estimators':[300],
              'reg_lambda': [1e-5, 1e-2, 0.1, 1, 100] ,
              'learning rate':[0.1]}

#GridSearchCV算是一個meta-estimator，參數中帶有estimator，像是SVC，重點是會創造一個新的estimator，但又表現的一模一樣。也就是estimator=SVC時，就是作為分類器
#Verbose可設定為任一整數，它只是代表數字越高，文字解釋越多
grid = GridSearchCV(XGBClassifier(),param_grid,verbose=3)

#利用剛剛設定的參數來找到最適合的模型
grid.fit(train_x,train_y)

#顯示最佳參數組合
grid.best_params_

#顯示最佳estimator參數
grid.best_estimator_

#利用剛剛的最佳參數再重新預測測試組
grid_predictions = grid.predict(test_x)

#評估新參數的預測結果好壞
print('\n')
print(grid.best_params_)
print('\n')
#print(classification_report(test_y,grid_predictions))
```
